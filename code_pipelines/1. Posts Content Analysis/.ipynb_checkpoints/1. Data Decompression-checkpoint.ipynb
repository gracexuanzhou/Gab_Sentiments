```python
import zstandard

def decompress_zstandard_to_folder(input_path, output_path):
    with open(input_path, 'rb') as compressed:
        decomp = zstandard.ZstdDecompressor()
        with open(output_path, 'wb') as destination:
            decomp.copy_stream(compressed, destination)

for i in range (1, 863): # 863 chunks
    file_n = str(i).zfill(5)
    decompress_zstandard_to_folder("/dbfs/mnt/lsde/gab/part-" + file_n + ".zst", "/dbfs/mnt/group12/data/part-" + file_n + ".bin")
```
